```python
import pandas as pd

student = {
    "name":["mandip","dip","rushal","pandey","dikshit","Dikshit"],
    "age":[22,21,20,21,23,23],
    "score":[85,88,87,95,90,90]
}

print(type(student))
```

```python
df = pd.DataFrame(student,index=['a','b','c','d','e','f'])
print("<--------------DATA--------------->")
print(df)
print("-"*70)
print(df.loc['a'])
```

```python
#max,min,sum,describe

print("max from whole table:\n")
print(df.max(), "\n")
print("max age from whole table:",df["age"].max())
print("-"*70)



print("minimum value from whole table:")
print(df.min(),"\n")
print("minimum score from whole table: ",df['score'].min())
print("-"*70)


print("whole sum:")
print(df.sum())
print("sum of marks of each student: ", df['score'].sum())
print("-"*70)


```
