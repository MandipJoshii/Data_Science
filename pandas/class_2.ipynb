# outliners

$$Z = \frac{X - \mu}{\sigma}$$

```python
students = {
    "marks":[86,70,57,56,71,1000]
}

df = pd.DataFrame(students)

student_mean = df['marks'].mean()
student_sd = df['marks'].std()

print(f"student mean: {student_mean:.2f}")
print(f"student sd: {student_sd:.2f}")
print("-"*50)
upper_limit = student_mean + (3 * student_sd)
lower_limit = student_mean - (3 * student_sd)

print(f"upper limit: {upper_limit:.2f}")
print(f"lower limit: {lower_limit:.2f}")



#determining outliner

cleaning_outliner = df[
    (df['marks'] > lower_limit) | 
    (df['marks'] < upper_limit)
    ]

print(cleaning_outliner)
    
```

```python
"""Z SCORE: TELL HOW FAR THE VALUE IS FROM MEAN USING MEAN AND SD Z > 3 IS AN OUTLINER, Z SCORE DEPENDS HEEAVILY ON MEAN """
```
