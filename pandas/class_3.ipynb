### EDA


```python
from sklearn.datasets import fetch_california_housing
import pandas as pd

california = fetch_california_housing(as_frame = True)

# print(california)
# print(type(california))

df = pd.DataFrame(california.data, columns=california.feature_names)

print("ORIGINAL DATA: ")
print(df.shape)
print("-"*100)

#formula = z = x - mean/sd



mean_income = df['MedInc'].mean()
print(f"mean of income: {mean_income:.2f}")
print("-"*100)

sd_income = df['MedInc'].std()
print(f"sd of income: {sd_income:.2f}")
print("-"*100)



print("Z SCORE: ")
df['z_score'] = (df['MedInc'] - mean_income) /sd_income

print(df[['MedInc','z_score']].head())
print("-"*100)

# print(df['z_score'].describe())
# print(df['z_score'].abs().head())
# print(df['z_score'].shape)
print("absolute value:")
df['abs_z_score'] = df['z_score'].abs()
print(df['abs_z_score'].head())











# print("3 sigma rule")
# print("-"*100)


# upper_limit = mean_income + (3*sd_income)
# lower_limit = mean_income - (3*sd_income)

# print(f"upper limit: {upper_limit:.2f}")
# print(f"lower limit: {lower_limit:.2f}")
# print("-"*100)


# print("removing outliners: ")

# df_clean = df[
#     (df['MedInc'] > lower_limit) &
#     (df["MedInc"] < upper_limit)
# ]

# print(df_clean.head())



```

```python
df_ranked = df.sort_values(by='abs_z_score', ascending=False)
print(df_ranked['abs_z_score'])



# df = df.sort_values(by='z_score', ascending=False)
# print(df.head())
```

```python
#normal = less than 1
#unusual = -1.5 to 1.5
# extreme = -3 to 3


df_ranked['label'] = 'normal'
# print(df_ranked)
df_ranked.loc[
    (df_ranked['abs_z_score'] > 1.5) | (df_ranked['abs_z_score'] < -1.5), 'label'
] = 'unusual'


df_ranked.loc[
    (df_ranked['abs_z_score'] > 3) | (df_ranked['abs_z_score'] < -3), 'label'
] = 'extreme'
print(df_ranked.sample(40))
```

## MIN-MAX 

y = x - min/max-min

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
cali = fetch_california_housing(as_frame=True)

print("-"*100)
print("DATA TYPE: ")
print(type(cali)) # bunch
print("-"*100)

```

```python
print("original data: ")
print(cali)
print("-"*100)

```

```python
print("original data but in dataframe:")
df = pd.DataFrame(cali.data, columns=cali.feature_names)
print(df.head())
print("-"*100)
```

```python
print("finding min_max: ")
df['min_max'] = (df['MedInc'] - min(df['MedInc']))/(max(df['MedInc'])-min(df['MedInc']))
print(df[['MedInc','min_max']].sample(20))
```

# z_index

```python
z_mean = df['MedInc'].mean()
print('mean: ',z_mean)

z_sd = df['MedInc'].std()
print('standard deviation: ',z_sd)
print("z_score: ")
df['z_score'] = (df['MedInc']-z_mean)/z_sd
print(df['z_score'].sample(20))
```
